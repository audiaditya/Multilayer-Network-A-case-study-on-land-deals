```python
 #importing libraries
import networkx as nx
import pylab
import matplotlib.pyplot as plt
import csv
import pandas as pd
```

```python
#loading the gold.net csv file 
gold_net = nx.DiGraph()
with open('deal_gold_network.csv') as f:
    f.readline()
    for line in f:
        s,d,w = line.lstrip().split(',')
        gold_net.add_edge(s,d,weight = int(w))
```

```python
#viewing total no.of nodes and edges in gold file
print("Nodes of graph: ")
print(gold_net.nodes())
print("Edges of graph: ")
print(gold_net.edges())
```

```python
#nx.draw which is NetworkX provides basic functionality for visualizing graphs.
nx.draw(gold_net, with_labels=True,node_color='g')
plt.show()
```

```python
#count of nodes
gold_net.number_of_nodes()
```

```python
#count of edges
gold_net.number_of_edges()
```

```python
gold_net.size()
```

```python
print(nx.info(gold_net))
```

```python
plt.figure(figsize=(10,10))
nx.draw_networkx(gold_net,node_color='g')
plt.show()
plt.savefig("Gold deal image ")
```

```python
print(nx.is_directed(gold_net)) 
```

```python
 #density
density = nx.density(gold_net)
print("Network density:", density)
#1 would indicate that all possible edges are present (a perfectly connected network).
```

```python
nx.degree(gold_net)
```

```python
def plot_degree_dist(gold_net):
    degrees = [gold_net.degree(n) for n in gold_net.nodes()]
    plt.hist(degrees)
    plt.show()

plot_degree_dist(nx.gnp_random_graph(100, 0.5, directed=True))
```

```python
#Degree centrality 
nx.degree_centrality(gold_net)
```

```python
most_influential= nx.degree_centrality(gold_net)
```

```python
for w in sorted(most_influential, key= most_influential.get, reverse=True): 
  print(w, most_influential[w])
```

```python
#betweenness centrality
nx.betweenness_centrality(gold_net)
#We see that BRA and COD countries are best connector
```

```python
#closeness centrality
close_centrality = nx.closeness_centrality(gold_net)
close_centrality
```
